```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]],columns=['cgpa','resume_score','package'])
df
```

```python
weights=initialize_weights([2,2,1])
lr=0.001
epoch=800
loss=[]
x,y=df.iloc[:,0:2].values,df.iloc[:,2].values
for i in range(epoch):
    for j in range(len(x)):
      X=x[j].reshape(2,1)
      Y=y[j].reshape(1,1)
      y_hat,a1=forward_feed(weights,X)
      update_weights(X,Y,y_hat,a1,weights)
      loss.append((y_hat-Y)**2)
    print('Epoch - ',i+1,'Loss - ',np.array(loss).mean())

print(weights)
```

```python
def initialize_weights(dim):
  np.random.seed(0)
  l=len(dim)
  weights={}
  for i in range(1,l):
    weights[f'w{i}']=np.ones((dim[i-1],dim[i]))*0.1
    weights[f'b{i}']=np.zeros((dim[i],1))
  return weights
```

```python
def forward_feed(weights,x):
  l=len(weights)//2
  a=x
  for i in range(1,l+1):
    w=weights['w'+str(i)]
    a_prev=a
    a=np.dot(w.T,a_prev)+weights[f'b{i}']
  return a,a_prev
```

```python
def update_weights(x,y,y_hat,a1,weights):
  weights['w1'][0][0]=weights['w1'][0][0]+lr*2*(y-y_hat)*(weights['w2'][0][0])*x[0][0]
  weights['w1'][0][1]=weights['w1'][0][1]+lr*2*(y-y_hat)*(weights['w2'][1][0])*x[0][0]
  weights['b1'][0][0]=weights['b1'][0][0]+lr*2*(y-y_hat)*(weights['w2'][0][0])

  weights['w1'][1][0]=weights['w1'][1][0]+lr*2*(y-y_hat)*(weights['w2'][0][0])*x[0][0]
  weights['w1'][1][1]=weights['w1'][1][1]+lr*2*(y-y_hat)*(weights['w2'][1][0])*x[1][0]
  weights['b1'][1][0]=weights['b1'][1][0]+lr*2*(y-y_hat)*(weights['w2'][1][0])

  weights['w2'][0][0]=weights['w2'][0][0]+lr*2*(y-y_hat)*a1[0][0]
  weights['w2'][1][0]=weights['w2'][1][0]+lr*2*(y-y_hat)*a1[1][0]
  weights['b2'][0][0]=weights['b2'][0][0]+lr*2*(y-y_hat)
```
