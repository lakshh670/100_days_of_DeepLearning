```python
# In this notebook we will se embeddings, a neural network technique which is used to vectorize the text. It is better the tokenizing and padding the text vectors because unlike padding it generates dense vectors and it also captures the sematic meaning(context) of the words.
# Initial steps are same, we need to tokenize and then do the padding of the text vectors
```

```python
import numpy as np

docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.index_word)
```

```python
sequences=tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequences=sequences,padding='post')
sequences
```

```python
from keras import Sequential
from keras.layers import Embedding, Dense,SimpleRNN,Flatten
```

```python
model = Sequential()
model.add(Embedding(input_dim=18,output_dim=2,input_length=5))

model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
# Here we are not training the model. Just doing prediction on initial weights
pred=model.predict(sequences)
print(pred)
```

```python
# Now lets build a simple RNN
from keras.datasets import imdb
(x_train,y_train),(x_test,y_test) = imdb.load_data()
```

```python
x_train,x_test=pad_sequences(x_train,padding='post',maxlen=50),pad_sequences(x_test,padding='post',maxlen=50)
```

```python
np.max(x_train)
```

```python
len(x_train)
```

```python
len(x_train[0])
```

```python
model = Sequential()
model.add(Embedding(np.max(x_train)+1, 2,input_length=50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(x_train, y_train,epochs=5,validation_data=(x_test,y_test))
```
