```python
# In this notebook we will see the different weight initialisation techniques. These are of following types.
# 1. Xavier or glorot (tanh, sigmoid):
#    a. Uniform
#    b. Normal
# 2. He(relu)
#    a. Uniform
#    b. Normal
# You can see the formulaes in the notes
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from matplotlib import pyplot as pyplot
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
```

# Data Generation

```python
x,y=make_moons(noise=0.12,random_state=42)
plt.scatter(x[:,0],x[:,1],c=y)
```

## Xavier_normal initialisation

```python
# We are doing this in keras, if we want to implement it in the keras then we can randomly initialise the weights through np.random.randn() and multiply it with the constant(see notes)
model = Sequential()

model.add(Dense(10,activation='tanh',input_dim=2,kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['Accuracy'])
history=model.fit(x,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y,clf=model)
```
