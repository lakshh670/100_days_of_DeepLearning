```python
# In this module we will build the LeNet CNN architecture
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,AvgPool2D,Flatten
```

```python
model=Sequential()
model.add(Conv2D(6,(5,5),input_shape=(32,32,1),activation='tanh',padding='valid')) # Convolution layer having 6 filters without any padding and strides.
# The resultant feature map will be of size (28,28,6)
model.add(AvgPool2D(pool_size=(2,2),strides=2,padding='valid')) # Avg pooling layer.
# Resultant f_map -->(14,14,6)

# These 2 layers combine to form the 1st layer
model.add(Conv2D(16,(5,5),activation='tanh',padding='valid'))
# Resultant f_map -->(10,10,16)
model.add(AvgPool2D(pool_size=(2,2),strides=2,padding='valid'))
# Resultant f_map -->(5,5,16)

# This is the 2nd layer
model.add(Flatten())
model.add(Dense(120,activation='tanh'))# 3rd layer
model.add(Dense(84,activation='tanh'))# 4th layer
model.add(Dense(10,activation='softmax'))# 5th layer. When we are working on the mnist digit dataset then we have 10 classes in the output, that's why the output layer has 10 nodes and the activation is softmax instead of binary cross entropy
# This a 5 layer architecture that's why its also called LeNet5.
```
